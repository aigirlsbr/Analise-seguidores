## Análise por mês (Maio - Julho)

```python
def nome_mes(day):
    if day == 1:
        return 'Janeiro'    
    elif day == 2:
        return 'Fevereiro'
    elif day == 3:
        return 'Março' 
    elif day == 4:
        return 'Abril' 
    elif day == 5:
        return 'Maio' 
    elif day == 6:
        return 'Junho'
    elif day == 7:
        return 'Julho'
    elif day == 8:
        return 'Agosto'
    elif day == 9:
        return 'Setembro'
    elif day == 10:
        return 'Outubro'
    elif day == 11:
        return 'Novembro'
    elif day == 12:
        return 'Dezembro'
```

```python
def line_day(data):
    mes = nome_mes(data['mes'].unique()[0])
    sns.set_palette('Purples_r')
    sns.set_style('darkgrid')
    ax = sns.lineplot(x='dia',y='Total de seguidores',data=data)
    ax.figure.set_size_inches(12,6)
    ax.set_title('Novos seguidores da comunidade no linkedin no mês {}'.format(mes),loc='left',fontsize=18)
    ax.set_xlabel('Dias',fontsize=14)
    ax.set_ylabel('Quantidade de novos seguidores',fontsize=14)
    ax=ax
```

```python
fevereiro = df[df['mes']==2]
line_day(fevereiro)
```

```python
marco = df[df['mes']==3]
line_day(marco)
```

Análises:
    1. Tivemos uma quantidade de seguidores menores que o mês de Fevereiro
    2. Tivemos um pico no dia 04 com 30 novos seguidores.
    3. Foi caindo a quantidade de seguidores e voltamos a ter um pico no dia 07, essa quantidade foi caindo ao decorrer do mês e só voltamos a ter novos seguidores no dia 16 com 12 novos seguidores

## O ganho de novos seguidores em uma determinada semana está ligado ao ganho da semana passada?

Poderemos descobrir isso utilizando a função de Autocorrelação. Ela é uma medida de correlação entre as observações de uma série temporal. Para entender melhor o funcionamento dela, comparamos os valores com suas próprias correlações, ou seja, consigo mesmas.




```python
from pandas.plotting import autocorrelation_plot
```

```python
ax = autocorrelation_plot(df['Total de seguidores'])
ax.figure.set_size_inches(12,6)
ax.set_title('Correlação do ganho de novos seguidores',loc='left',fontsize=18)
ax=ax
```

```python
ax = autocorrelation_plot(df_mes['aumento'][1:])
ax.figure.set_size_inches(12,6)
ax.set_title('Correlação do aumento de novos seguidores',loc='left',fontsize=18)
ax=ax
```

## Normalizar o ganho de seguidores na semana (retirando os finais de semana)

```python
qtd_dias_final_semana ={'fevereiro':10,'março':9,'abril':8,'maio':10,'junho':8,'julho':8}
```

```python
final_semana = pd.DataFrame(qtd_dias_final_semana.items(), columns=['Mes', 'qtd_final_semana'])
final_semana
```

```python
#vou pegar a qtd de novos seguidores e dividir pelo dias que foram finais de semana
df_mes['novos_seguidores_normalizados'] = df_mes['Total de seguidores']/final_semana['qtd_final_semana'].values
df_mes.head(14)
```

```python
plt.figure(figsize=(16,14))
ax = plt.subplot(2,1,1)
ax.set_title('Análise dos novos seguidores normalizados no Linkedin',fontsize=18,loc='left')
sns.lineplot(x='mes',y='novos_seguidores_normalizados',data=df_mes)
ax = plt.subplot(2,1,2)
sns.lineplot(x='mes',y='aumento',data=df_mes)
ax=ax
```

Analisar sempre tendência, sazonalidade e ruído.

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
resultados = seasonal_decompose(df['Total de seguidores'],freq=3)
ax = resultados.plot()
```

```python
observacao = resultados.observed
tendencia = resultados.trend
sazonalidade = resultados.seasonal
ruido = resultados.resid
```

```python
data = ({
       'observacao':observacao,
       'tendencia':tendencia, 
       'sazonalidade':sazonalidade,
       'ruido':ruido
})
resultado = pd.DataFrame(data)
resultado.head
```
